```python
# 01_data_cleaning.ipynb

import pandas as pd
import os

# Set path to dataset
DATA_PATH = "../data/sample_sales.csv"

# Load the dataset
df = pd.read_csv(DATA_PATH)

# Preview
print("Initial rows:")
display(df.head())

# Data types
print("Data types:")
print(df.dtypes)

# Handle missing values (example strategy)
print("Missing values:")
print(df.isnull().sum())

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows with missing values (or handle differently)
df_clean = df.dropna()

# Create additional columns if needed (e.g., Month)
df_clean['Month'] = df_clean['Date'].dt.to_period('M')

# Save cleaned version
CLEANED_PATH = "../data/cleaned_sales.csv"
df_clean.to_csv(CLEANED_PATH, index=False)

print(f"Cleaned data saved to: {CLEANED_PATH}")
```
